```python
# linking with SparkSession
from pyspark.sql import SparkSession
spark = SparkSession.builder.master("local[*]").appName('Visualization used in the paper').getOrCreate()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame({'Hyperparameter':['Initial ( 0.62 ) ', 'Tuned ( 0.9 ) ',], 'Accuracy':[0.62, 0.9]})

ax = df1.plot.bar(x='Hyperparameter', y='Accuracy', rot=0,figsize=(10,10))

index=['Accuracy']



plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.xlabel('Hyperparameter', fontsize=18,color='red')
plt.ylabel('Accuracy of Random Forest classifier', fontsize=18,color='red')
plt.legend(index,loc="upper center",fontsize=10,edgecolor='red')
plt.show()
```

```python
df2 = pd.DataFrame({'Machine Learning classifiers':['Decision Tree(0.62)', 'Logistic Regression(0.72)','Naive Bayes(0.42)','Random Forest(0.9)'], 'Accuracy':[0.6191780821917808, 0.7276497695852534,0.42811059907834104,0.9]})

ax = df2.plot.barh(x='Machine Learning classifiers', y='Accuracy',rot=0,figsize=(10,10),color=["green","blue","black","cyan"])

index=['Decision Tree','Logistic Regression','Naive Bayes','Random Forest']


plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylabel('Machine Learning classifiers', fontsize=18,color='red')
plt.xlabel('Accuracy ', fontsize=18,color='red')
plt.savefig("compare.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

#create data dictionary
df3 = pd.DataFrame({'Classification Result': ['Accuracy', 'F1-Score', 'Precision', 'Recall'],
                   'Random Forest': [ 0.9, 0.912037037037037 ,0.9078341013824884 ,  0.9162790697674419],
                   'Decision Tree': [ 0.6191780821917808, 0.6744186046511628 	,  0.6744186046511628 , 0.6744186046511628 ],
                   'Naive Bayes': [0.42811059907834104 , 0.3143507972665148, 0.2875 , 0.34673366834170855],
                   'Logistic Regression': [0.7276497695852534, 0.7980997624703089  ,0.7567567567567568   , 0.8442211055276382]})

index=['Random Forest','Decision Tree','Naive Bayes','Logistic Regression']


#plot columns on bar chart
df3[['Classification Result', 'Random Forest','Decision Tree','Naive Bayes','Logistic Regression']].plot(x='Classification Result', kind='bar',figsize=(12,12))
plt.xticks(fontsize=16,rotation=45)
plt.yticks(fontsize=16)
plt.xlabel('Classification  Result', fontsize=18,color='red')
plt.ylabel('Multiple classifiers', fontsize=18,color='red')
plt.title('Comparative analysis of different classifiers supported by Apache Spark',fontsize=20,color='red')
plt.legend(index,loc="upper center",fontsize=10,edgecolor='red')
plt.savefig("classifiers.png")
plt.show()
```
